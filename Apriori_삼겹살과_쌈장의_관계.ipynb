```python
# Apriori 분석에 필요한 라이브러리 설치
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = [
           ['삼겹살', '쌈장', '콜라'],
           ['항정살', '상추'],
           ['사이다', '콜라'],
           ['삼겹살', '쌈장', '사이다'],
           ['삼겹살', '쌈장', '상추'],
           ['항정살', '쌈장', '상추'],
           ['삼겹살', '항정살', '쌈장', '상추', '콜라']
]
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset). transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_) #array를 보기좋게 DF로 변경
```

```python
df
```

```python
# 지지도(Support) 최소지지도 = 0.5
frequent_itemsets = apriori(df, use_colnames=True)
frequent_itemsets
```

```python
apriori(df, min_support=0.3, use_colnames=True).sort_values(by='support', ascending=False)
```

```python
# 신뢰도(Confidence) 와 향상도(lift)
association_rules(frequent_itemsets, metric="confidence") 
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3) 
# antecedent support & consequent support 뭘까?
# 아! 알아냄 >> 
# 첫번째 행을 기준으로 각 선행으로 사건(삼겹살), 이후에 일어난 사건(쌈장)의 지지도이다.
# support는 {삼겹살, 쌈장}의 집합 지지도
```
